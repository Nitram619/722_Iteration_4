```python
import findspark
findspark.init('C:\spark\spark-3.1.2-bin-hadoop3.2')
import pyspark
import pandas
import sklearn
import numpy as np
import matplotlib.pyplot as plt
from pyspark.sql import SparkSession, DataFrame, DataFrameNaFunctions, Column, SQLContext
import pyspark.sql.functions as fun
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.regression import LinearRegression
from pyspark.ml.regression import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from pyspark.ml.feature import VectorAssembler
import pandas as pd
import sklearn
from sklearn.preprocessing import PowerTransformer
from pandas import DataFrame
import matplotlib.pyplot
from pyspark.ml.feature import ChiSqSelector
from pyspark.sql import SQLContext
from sklearn.metrics import explained_variance_score


spark = SparkSession.builder.appName('Iteration_4').getOrCreate()
df1 = spark.read.csv('covid19_italy.csv', inferSchema=True, header = True)
df2 = spark.read.csv('Italy_Part2.csv', inferSchema=True, header = True)
df = df1.union(df2)
#df.printSchema()
#df.show()
df.count()
# Step 2.3

#3.1 Select the data
drop_columns = ['SNo', 'RegionCode', 'Latitude', 'Longitude', 'Month', 'Date', 'Country', 'RegionName']
df = df.drop(*drop_columns)
#df.show()
#df.printSchema()


df = df.na.fill(0)
#df.show()
#df.describe().show()

new_df = df.withColumn('Target', df['Deaths'])
new_df.agg({"Target":"mean"}).show()
pdata = new_df.toPandas()

#new_df.show()
#new_df.describe().show()
new_df.printSchema()


#4.1 Reduce the data
panda_data = new_df.toPandas()

print('Correlation: \n', panda_data.corr()['Target'])
newd = new_df.toPandas()
new_data = newd.values[:, :-1]
pt = PowerTransformer(method='yeo-johnson')
new_data = pt.fit_transform(new_data)

dataset1 = DataFrame(new_data)

dataset1.hist()

vectorassembler = VectorAssembler(inputCols = [ 'HospitalizedPatients', 'IntensiveCarePatients', 'TotalHospitalizedPatients', 'HomeConfinement', 
                                               'CurrentPositiveCases', 'NewPositiveCases', 'Recovered', 'TotalPositiveCases',
                                               'TestsPerformed'], outputCol = 'features')
output = vectorassembler.transform(new_df)
finaloutput = output.select('features', 'Deaths')
#output.printSchema()
train_data,test_data = finaloutput.randomSplit([0.7,0.3])
train_data.describe().show()
test_data.describe().show()

lr = LinearRegression(featuresCol='features', labelCol='Deaths', predictionCol='prediction')
lrmodel = lr.fit(train_data)
print("Coefficients: " + str(lrmodel.coefficients))
print("Intercept: " + str(lrmodel.intercept) + "\n")
trainingSummary = lrmodel.summary
print("RMSE: %f" % trainingSummary.rootMeanSquaredError)
print("r2: %f" % trainingSummary.r2)

lrprediction = lrmodel.transform(test_data)
lrprediction.select('prediction', 'Deaths').show(9)



lrevaluator = RegressionEvaluator(predictionCol='prediction', \
                 labelCol='Deaths',metricName="r2")
test_results = lrmodel.evaluate(test_data)
print('Model: linear regression')
print("rootMeanSquaredError: {}".format(test_results.rootMeanSquaredError))
print("meanSquaredError: {}".format(test_results.meanSquaredError))
print("meanAbsoluteError: {}".format(test_results.meanAbsoluteError))
#print("Explained variance: {}".format(test_results.explainedVariance))
print("R2: {}".format(test_results.r2))
test_results.residuals.show(9)
columns = [ 'HospitalizedPatients', 'IntensiveCarePatients', 'TotalHospitalizedPatients', 'HomeConfinement', 
                                               'CurrentPositiveCases', 'NewPositiveCases', 'Recovered', 'TotalPositiveCases',
                                               'TestsPerformed']
print("Intercept: {}".format(lrmodel.intercept))
dia = pd.DataFrame({'attributes': columns, 'coefficient': lrmodel.coefficients,})
print(dia)
lrarr = np.array(test_results.predictions.select('Deaths').collect())
lrpredictionarr = np.array(test_results.predictions.select('Prediction').collect())
plt.figure(figsize=(10,8))
plt.scatter(lrarr,lrpredictionarr, alpha=.75,color='r') 
 
plt.xlabel('Deaths')
plt.ylabel('prediction')
plt.title('Linear Regression Model')

plt.show()





```

```python
rf = RandomForestRegressor(labelCol='Deaths', featuresCol='features')
rfmodel = rf.fit(train_data)
print('Random forest Regressor\n')
rftrainresults = rfmodel.transform(train_data)
rftrainresults.select('Deaths','prediction').show(10)
rfpredictions = rfmodel.transform(test_data)
rfevaluator = RegressionEvaluator(labelCol='Deaths', predictionCol='prediction')
rfpredictions.select('Deaths','prediction').show(10)

print('Mean Absolute Error = %g' %rfevaluator.evaluate(rftrainresults,{rfevaluator.metricName: "mae"}))
print('Mean Squared Error = %g'%rfevaluator.evaluate(rftrainresults,{rfevaluator.metricName: "mse"}))
print('Root Mean Squared Error = %g' %rfevaluator.evaluate(rftrainresults,{rfevaluator.metricName: "rmse"}))
print('Random forest Regressor')
print('rootMeanSquaredError:', rfevaluator.evaluate(rfpredictions, {rfevaluator.metricName: "rmse"}))
rfpredictions = rfpredictions.toPandas()
plt.plot(rfpredictions['Deaths'],rfpredictions['prediction'], 'o', color='red') 
plt.xlabel('Actual Value of Deaths value')
plt.ylabel('Predicted Value of Deaths value')
plt.title('Random Forest')
plt.show()
```
